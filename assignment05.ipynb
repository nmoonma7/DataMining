```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from skimage import io, color
from skimage import exposure
```

# Function for computing the magnitude of the gradient : 

```python
def computeMagnitude(vecOne,vecTwo) : 
    return np.sqrt(vecOne**2+vecTwo**2)
```

# Function for computing the direction of the gradient :

```python
def computeDirection(vecOne,vecTwo) : # 두 벡터간의 거리를 계산한다
    return np.arctan(vecTwo/vecOne)
```

```python
file_image = 'cau.jpg'

im_color = io.imread(file_image)
im_gray  = color.rgb2gray(im_color)

ker = np.array([[0,0,0],[0,1,0],[0,0,0]]) 
im_conv= signal.convolve2d(im_gray, ker, boundary='symm', mode='same') 

p1 = plt.subplot(2,2,1)
p1.set_title('color image')
plt.imshow(im_color)
plt.axis('off')

p2 = plt.subplot(2,2,2)
p2.set_title('gray image')
plt.imshow(im_gray, cmap='gray')
plt.axis('off')

p3 = plt.subplot(2,2,3)
p3.set_title('convolution kernel')
plt.imshow(ker, cmap='gray')
plt.axis('off')

p4 = plt.subplot(2,2,4)
p4.set_title('convolution result')
plt.imshow(im_conv, cmap='gray')
plt.axis('off')

plt.show()
```

# derivative in x-direction

```python
# derivative in x-direction
ker_derX = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
im_derX= signal.convolve2d(im_gray, ker_derX, boundary='symm', mode='same') 

plt.imshow(im_derX, cmap='gray')
plt.axis('off')
```

# derivative in y-direction

```python
# derivative in y-direction

ker_derY = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
im_derY= signal.convolve2d(im_gray, ker_derY, boundary='symm', mode='same') 

plt.imshow(im_derY, cmap='gray')
plt.axis('off')
```

# absolute value of gradient

```python
# computing the magnitude of the gradient
#mag_grad=np.sqrt(im_derX**2+im_derY**2)
mag_grad=computeMagnitude(im_derX,im_derY) 

plt.imshow(mag_grad, cmap='gray')
plt.axis('off')
```

# direction of gradient

```python
# computing the direcetion of the gradient
# tan^-1(derivative_y/derivative_x)

# dir_grad=np.arctan(im_derY/im_derX)
dir_grad=computeDirection(im_derX,im_derY)

plt.imshow(dir_grad, cmap='hsv')
plt.axis('off')
```

# result image with smoothing kernel

```python

ker_smooth = np.array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
im_smooth= signal.convolve2d(im_gray, ker_smooth, boundary='symm', mode='same') 

plt.imshow(im_smooth, cmap='gray')
plt.axis('off')
```

# result image with your own kernel : sharpen kernel

```python

ker_sharp = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
im_sharp= signal.convolve2d(im_gray, ker_sharp, boundary='symm', mode='same') 

plt.imshow(im_sharp, cmap='gray')
plt.axis('off')
```
